```python
import os
mainpath = './work_dirs/'
def remove_early_model(dirName, removeEpoch):
    for index in removeEpoch:
        filename = 'epoch_' + str(index) + '.pth'
        model_Path = os.path.join(mainpath, dirName, filename)
        try:
            os.remove(model_Path)
            print(f"{model_Path} 삭제!")
        except:
            print(f"{model_Path}File not exist.")
```
